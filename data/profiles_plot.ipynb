```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('all_profiles.csv')
```

```python
results
```

```python
# Plot all runs VTG capacity in a graph.
sns.set_theme(style="dark")
g = sns.relplot(
    data=results,
    x="t", y="total_VTG_capacity", hue="sample_id",
     kind="line", palette="crest", linewidth=0.01, legend=False,
)
```

```python
kpis = results.columns.drop(['sample_id', 't'])


g = sns.relplot(
data=results,
x="t", y=kpis[0], hue="sample_id",
    kind="line", palette="crest", linewidth=0.01, legend=False,
)
```
